```python
# raw HTML content
import requests

# webURL = "https://anmol1804.github.io/MyPortfolio/"
webURL = "https://github.com/dyte-in/javascript-sample-app"
r = requests.get(webURL)
print(r.content)
```

```python
# raw HTML content using BeautifulSoup
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.content,'html.parser')
print(soup.prettify())
```

```python
# list of a tags
a_list = soup.find_all('a')
print(a_list)
```

```python
# text of a tags

print("Text from a tags : \n")
for a_tag in a_list:
    print(a_tag.get_text())
```

```python
# list of p tags
p_list = soup.find_all('p')
print(p_list)
```

```python
# text of p tags
for p_tag in p_list:
    print(p_tag.get_text())
```

```python
# list of div tags
div_list = soup.find_all('div')
print(div_list)
```

```python
# text of div tags
for div_tag in div_list:
    print(div_tag.get_text())
```

```python
title = soup.find('title')
print(title.get_text())
```

```python
# All text
text = soup.get_text()
print(text)
```

```python
# text after removing puunctuations
webURL = "https://anmol1804.github.io/MyPortfolio/"
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

r = requests.get(webURL)

soup = BeautifulSoup(r.content,'html.parser')
text = soup.get_text()

for ele in text:
    if ele in punc:
        text = text.replace(ele, "")
        
print(text)
```
